```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt

# Set Plot Style
plt.style.use('seaborn-paper')
```

```python
# Make a Plot
plt.clf()
fig, ax = plt.subplots(2, sharex = True, figsize = (20,16))

# Import the results of Woodberry - Ideal
Data_Ideal = pd.read_csv('../Modelica/Rosenbrock_Ideal.csv')
# Use the steady simulation state from t = 2400...3600
Data = Data_Ideal[((Data_Ideal['time']>399.9) & (Data_Ideal['time']<600.1))]
# Get the values
time = Data['time'].values
y1 = Data['G_11.y'].values
y2 = Data['G_22.y'].values
u1 = Data['r_11.y'].values
u2 = Data['r_22.y'].values
IAE1 = Data['IAE_1.y'].values
IAE2 = Data['IAE_2.y'].values
IE1 = Data['IE_1.y'].values
IE2 = Data['IE_2.y'].values
# Set IAE to zero
IAE1 = IAE1-IAE1[0]
IAE2 = IAE2-IAE2[0]

# Make a Plot
plt.clf()
fig, ax = plt.subplots(2, sharex = True, figsize = (10,8))

# y1
ax[0].plot(time, y1)
# y2
ax[1].plot(time,y2, label = "Ideal System")
```
